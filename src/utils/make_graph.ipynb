```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import config

age_groups = config.AGE_GROUPS
class_names = config.GROUP_NAMES
# need '\n' in each class name so that it fits better in the plot 
for cnt, name in enumerate(class_names):
    class_names[cnt] = name.replace('-', '-\n')
```

```python
# manually copied individual AUC-scores here
xgboost_auc = np.array([0.68,0.70,0.64,0.50,0.63,0.50,0.81,0.84,0.76,0.76,0.91,0.99,0.93,0.97,0.94])
xresnet50_auc = np.array([0.63,0.66,0.51,0.61,0.59,0.53,0.71,0.81,0.80,0.86,0.96,0.90,0.62,0.95,0.88])

# do plotting
barwidth = 0.4
plt.bar(age_groups+barwidth, xgboost_auc, width=barwidth)
plt.xticks(range(len(class_names)), class_names, size=9)

plt.bar(age_groups, xresnet50_auc, width=barwidth)
plt.xlabel("Age-group")
plt.ylabel("Macro-AUC score")
plt.legend(["XGBoost", "XResNet50"])
plt.show()
```

```python
# load metadata-file
metadata = pd.read_csv(config.DATASET_METADATA_FILE)
# get age-bin distribution
age_bin_distribution = metadata[config.LABEL_COLUMN_NAME].value_counts()
age_bin_distribution
```

```python
# plot age-bin distribution
x = list(age_bin_distribution.keys())
y = list(age_bin_distribution.values)

shifted_class_names = ['']+class_names # so that names align with bars properly
plt.bar(x, y, color = "darkgreen")
plt.xticks(range(len(shifted_class_names)), shifted_class_names, size=9)
plt.xlabel('Age-group')
plt.ylabel('Number of subjects')
#plt.title('dataset age distribution')

plt.show()
```
